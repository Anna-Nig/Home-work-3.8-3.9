Задача 1. Фильтрация и трансформация данных (lambda, map, filter)
Дан список словарей, представляющих данные о студентах:
students = [
    {'name': 'Алексей', 'grade': 85, 'city': 'Москва'},
    {'name': 'Мария', 'grade': 92, 'city': 'Казань'},
    {'name': 'Иван', 'grade': 78, 'city': 'Москва'},
{'name': 'Елена', 'grade': 88, 'city': 'Иннополис'},
    {'name': 'Дмитрий', 'grade': 55, 'city': 'Москва'},
]
1. Используя filter и lambda, отфильтруйте студентов, чья оценка (grade) выше 80.
2. Используя map и lambda, преобразуйте отфильтрованный список, оставив только имена студентов в верхнем регистре.
3. Используя reduce (из functools) и lambda, вычислите среднюю оценку всех студентов из исходного списка. Результат округлите до двух знаков после запятой.

```python
from functools import reduce

students = [ 
    {'name': 'Алексей', 'grade': 85, 'city': 'Москва'}, 
    {'name': 'Мария', 'grade': 92, 'city': 'Казань'}, 
    {'name': 'Иван', 'grade': 78, 'city': 'Москва'}, 
    {'name': 'Елена', 'grade': 88, 'city': 'Иннополис'}, 
    {'name': 'Дмитрий', 'grade': 55, 'city': 'Москва'}
]

# 1. Фильтрация студентов с grade > 80
filtered_students = list(filter(lambda s: s['grade'] > 80, students))

# 2. Преобразование имён в верхний регистр
upper_names = list(map(lambda s: s['name'].upper(), filtered_students))

# 3. Вычисление средней оценки всех студентов через reduce
total_all = reduce(lambda x, y: x + y['grade'], students, 0)  
avg_all = total_all / len(students)

# Вывод результатов
print("Отфильтрованные студенты (grade > 80):")
for s in filtered_students:
    print(s)
print("\nИмена в верхнем регистре:")
print(upper_names)
print(f"\nСредняя оценка: {avg_all:.2f}") 
```

Задача 2. Функциональная обработка данных в Pandas
Дан DataFrame с информацией о продажах:
import pandas as pd

    'quantity': np.random.randint(1, 10, size=8)

1. ИспользуяmapдляSeries,создайтеновыйстолбец'revenue',которыйбудет равен price * quantity.
2. Сгруппируйте данные по продукту и вычислите общую выручку для каждого продукта.
 
3. Используя apply с lambda-функцией, отфильтруйте строки, где количество проданных единиц (quantity) больше среднего значения quantity по всему DataFrame. Выведите получившийся DataFrame.

```python
import pandas as pd
import numpy as np

np.random.seed(42)

data = {
    'product': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B'],
    'price': [100, 150, 100, 200, 150, 100, 200, 150],
    'quantity': np.random.randint(1, 10, size=8)
}
df = pd.DataFrame(data)


# 1. Создание столбца 'revenue' через map (умножение price * quantity)
df['revenue'] = (df['price'] * df['quantity']).map(lambda x: x)  


# 2. Группировка по продукту и общей выручке
revenue_by_product = df.groupby('product')['revenue'].sum()
print("Выручка по продуктам:")
print(revenue_by_product)

# 3. Фильтрация строк с quantity > среднего через apply с lambda
mean_quantity = df['quantity'].mean()
filtered_df = df[df.apply(lambda row: row['quantity'] > mean_quantity, axis=1)]
print("\nПродажи с quantity > среднего ({:.1f}):".format(mean_quantity))
print(filtered_df)
```

Напишите программу, которая выполняет следующие действия:
1. Определите функцию get_list, которая принимает строку из целых чисел, разделенных пробелами, и возвращает список из этих чисел.
2. Создайте функцию sort_func, которая:
a. Принимаетфункциюкакаргумент
b. Вызывает переданную функцию
c. Сортирует результат работы функции по возрастанию
d. Возвращаетотсортированныйрезультат
3. Используйте вызов функций get_list и sort_func, чтобы вывести результат на экран.

```python
# 1. Функция get_list: принимает строку из целых чисел, разделенных пробелами, и возвращает список из этих чисел.
def get_list(input_string):
    return list(map(int, input_string.split()))
    
# 2. Функция sort_func: принимает функцию, вызывает её, сортирует результат
def sort_func(func):
    def nested():  
        return sorted(func())
    return nested()
# 3. Использование
input_str = "3 1 4 1 5 9 2 6"
print("Исходный список:", get_list(input_str))
sorted_result = sort_func(lambda: get_list(input_str))
print("Отсортированный список:", sorted_result)
```

Задача 4. Декоратор для измерения времени выполнения
Реализуйте декоратор timer, который:
1. Замеряет время выполнения декорируемой функции
2. Выводит на экран сообщение: "Функция {имя_функции} выполнилась за
{время:.4f} сек."
3. Возвращаетрезультатвыполненияфункции
4. Корректно работает с любыми функциями (любое количество аргументов)
5. Сохраняет имя (__name__) и документацию (__doc__) исходной функции
Примечание: п.5 реализовать вручную или использовать functools.wrap, который копирует метаданные (__name__, __doc__) из исходной функции в функцию- обёртку
Примените этот декоратор к функции, которая вычисляет сумму чисел от 1 до 1_000_000, и выведите результат.

```python
import time
import functools

def timer(func):
    """Декоратор для измерения времени выполнения функции"""
    
    @functools.wraps(func) #  Сохраняет __name__, __doc__ автоматически
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs) # Вызов оригинальной функции
        end_time = time.time()
        print(f"Функция {func.__name__} выполнилась за {end_time - start_time:.4f} сек.")
        return result
    return wrapper

@timer
def sum_to_n(n):
    """Возвращает сумму чисел от 1 до n"""
    return sum(range(1, n + 1))

result = sum_to_n(1_000_000)
print(f"Результат: {result}")
```
