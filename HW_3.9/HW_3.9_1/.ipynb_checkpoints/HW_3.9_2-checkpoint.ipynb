Задача 2. Поиск и суммирование чисел через цепочку файлов
Для решения используйте многопроцессность с помощью метода pool.starmap.
Вам дан архив с 1000 текстовыми файлами, Архив с файлами 1 (path_8_8.zip). Задача заключается в том, чтобы написать код, который обрабатывает каждый из этих файлов в многопроцессном режиме.
В каждом файле записан путь к файлу в другом архиве. Архив с файлами 2 (recursive_challenge_8_8.zip). Ваш код должен следовать этому пути, чтобы найти конечный файл, содержащий число. Это число необходимо прибавить к глобальному счётчику.
Требования к коду:
1. Код должен открыть каждый текстовый файл из первого архива, считать путь, указанный внутри.
2. Перейти по указанному пути к целевому файлу и извлечь из него число.
3. Найденные числа необходимо суммировать.
4. Рекомендуется использовать конструкцию with
5. Для отчета по работе выполнить задание в файле ipynb или py. Обратить внимание на возможную разницу в работе при запуске Jupiter notebook и обычного файла py.


```python
import zipfile
import os
from multiprocessing import Pool

def process_file(args):
    """
    Обрабатывает 1 файл из первого архива:
    1. Читает путь из файла
    2. Идет по пути во втором архиве  
    3. Извлекает число из конечного файла
    """
    file1_path, zip1_path, zip2_path = args
    
    try:
        # 1. Читаем путь из файла первого архива
        with zipfile.ZipFile(zip1_path, 'r') as zip1:
            file_content = zip1.read(file1_path).decode('utf-8').strip()
        
        # 2. Идем по пути во втором архиве
        with zipfile.ZipFile(zip2_path, 'r') as zip2:
            final_content = zip2.read(file_content).decode('utf-8').strip()
        
        # 3. Извлекаем число из конечного файла
        number = int(final_content)
        return number
        
    except Exception as e:
        print(f"Ошибка в файле {file1_path}: {e}")
        return 0

def main():
    # Пути к архивам
    zip1_path = "path_8_8.zip"           # Архив 1 (1000 текстовых файлов)
    zip2_path = "recursive_challenge_8_8.zip"  # Архив 2 (целевая структура)
    
    # Получаем список всех файлов из первого архива
    with zipfile.ZipFile(zip1_path, 'r') as zip1:
        file_list = zip1.namelist()
    
    print(f"Найдено файлов в первом архиве: {len(file_list)}")
    
    # Создаем аргументы для starmap: [(file_path, zip1_path, zip2_path), ...]
    args_list = [(file_path, zip1_path, zip2_path) for file_path in file_list]
    
    # ✅ Многопроцессность с pool.starmap()
    with Pool(processes=10) as pool:  # 10 процессов
        results = pool.starmap(process_file, args_list)
    
    # Суммируем все найденные числа
    total_sum = sum(results)
    print(f"Общая сумма чисел: {total_sum}")

if __name__ == '__main__':
    main()
```
