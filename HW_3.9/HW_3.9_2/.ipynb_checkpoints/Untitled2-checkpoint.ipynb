Задача 2. Поиск и суммирование чисел через цепочку файлов
Для решения используйте многопроцессность с помощью метода pool.starmap.
Вам дан архив с 1000 текстовыми файлами, Архив с файлами 1 (path_8_8.zip). Задача заключается в том, чтобы написать код, который обрабатывает каждый из этих файлов в многопроцессном режиме.
В каждом файле записан путь к файлу в другом архиве. Архив с файлами 2 (recursive_challenge_8_8.zip). Ваш код должен следовать этому пути, чтобы найти конечный файл, содержащий число. Это число необходимо прибавить к глобальному счётчику.
Требования к коду:
1. Коддолженоткрытькаждыйтекстовыйфайлизпервогоархива,считатьпуть, указанный внутри.
2. Перейти по указанному пути к целевому файлу и извлечь из него число.
3. Найденныечисланеобходимосуммировать.
4. Рекомендуется использовать конструкцию with
5. Для отчета по работе выполнить задание в файле ipynb или py. Обратить внимание на возможную разницу в работе при запуске Jupiter notebook и обычного файла py.

```python
from worker import process_file
from multiprocessing import Pool
import zipfile
```

```python
def main():   # ОПРЕДЕЛЕНИЕ ПУТЕЙ К АРХИВАМ
    zip1_path = "path_8_8.zip"
    zip2_path = "recursive_challenge_8_8.zip" 

    
    # ПОЛУЧЕНИЕ СПИСКА ФАЙЛОВ ИЗ ПЕРВОГО АРХИВА
    # namelist() возвращает ['path/v2qadmwz.txt', 'path/v3l9excq.txt', ...]
    # Каждый файл содержит путь 
    with zipfile.ZipFile(zip1_path, 'r') as zip1: # Автоматическое закрытие
        files = zip1.namelist() 
        
    # ВЫВОД СТАТИСТИКИ
    # Показывает общее количество задач для обработки
    print(f" Файлов: {len(files)}")

    # ПОДГОТОВКА АРГУМЕНТОВ ДЛЯ multiprocessing.starmap()
    # starmap() распакует каждый кортеж 
    args_list = [(f, zip1_path, zip2_path) for f in files]

    # ПАРАЛЛЕЛЬНАЯ ОБРАБОТКА (10 процессов одновременно)
    # Каждый процесс: читает path/*.txt → находит число → возвращает int
    with Pool(processes=10) as pool:
        results = pool.starmap(process_file, args_list)

    # ФИНАЛЬНОЕ СУММИРОВАНИЕ (в главном процессе)
    # Python sum() над списком из 1001 чисел
    total_sum = sum(results)
    print(f" СУММА: {total_sum}")
    
main() # Запуск программы
```
